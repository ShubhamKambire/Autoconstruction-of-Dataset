```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
import os
import cv2
import matplotlib.pyplot as plt
```

```python
trangen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
                            )

training_set = trangen.flow_from_directory('main_data/Train',
                                           target_size=(64,64),
                                           batch_size=32,
                                           class_mode = 'binary')
```

```python
test_set = trangen.flow_from_directory('main_data/Test',
                                           target_size=(64,64),
                                           batch_size=32,
                                           class_mode = 'binary')
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128, activation ='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
cnn.compile(optimizer= 'adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
cnn.fit(x=training_set, validation_data= test_set, epochs =25)
```

```python
import os
import numpy as np
from keras.preprocessing import image
import cv2
n=0
path = "Data"
images1=[]

for i in os.listdir(path):
    
    test_image = image.load_img(str(path)+"/"+i, target_size = (64, 64))
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis = 0)
    result = cnn.predict(test_image)
    training_set.class_indices
    if result[0][0] == 1:
      prediction = 'BH'
      images1.append(i)
    else:
      prediction = 'FH'
    
```

```python
ss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()
```

```python
image_dir = []
n=0


for i in images1:
    im = cv2.imread(os.path.join(path,i))
    ss.setBaseImage(im)
    ss.switchToSelectiveSearchFast()
    pops = ss.process()
    
    for j in pops:
        x,y,w,h = j
        if x<1200:
            n = str(n)
            print(x,y,w,h)
            crp = im[x:x+w, y:y+h]
            resized = cv2.resize(crp, (64,64), interpolation = cv2.INTER_AREA)
            image_dir.append(resized)
    
    
    
    
```

```python
train_da = ImageDataGenerator(rescale = 1./255,
                             shear_range = 0.2,
                             zoom_range = 0.2,
                             horizontal_flip=True)
train_data = train_da.flow_from_directory('maine/train',
                                          target_size = (64,64),
                                         batch_size = 32,
                                         class_mode = 'binary')
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```
